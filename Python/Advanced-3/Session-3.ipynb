```python
import pandas as pd

df = pd.read_csv('owid-covid-data.csv')
df.set_index(pd.to_datetime(df.date), inplace=True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(14,4), dpi=200)
plt.boxplot(df.total_cases.dropna(), vert=False, showfliers=False, whis=[0,95], showmeans=True)
plt.show()
```

```python
country_name = 'Iran'

plt.figure(figsize=(14,4), dpi=200)
df[df.location==country_name].new_cases.plot()
plt.title(f'New Cases in {country_name}')
plt.xlabel('Date')
plt.ylabel('Cases(#)')
plt.show()
```

```python
# draw a plot for more than one country

plt.figure(figsize=(14,4), dpi=200)
df[df.location=='Iran'].new_cases.plot(label='Iran')
df[df.location=='Afghanistan'].new_cases.plot(label='Afghanistan')
plt.title(f'New Cases')
plt.xlabel('Date')
plt.ylabel('Cases(#)')
plt.legend()
plt.show()
```
